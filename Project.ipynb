```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
data = pd.read_csv("F:\\DEPI\\Datasets\\Datasets\\Superstore Sales Dataset.csv")
```

```python
df = data
```

```python
df
```

```python
print(df.info())
```

```python
df = df.iloc[:, :18]
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df["Postal Code"] = df["Postal Code"].fillna("Unknown")
```

```python
df = df.drop_duplicates()
```

```python
# Convert dates
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)
df['Ship Date'] = pd.to_datetime(df['Ship Date'], dayfirst=True)
```

```python
df['Delivery Time (Days)'] = (df['Ship Date'] - df['Order Date']).dt.days
df['Order Year'] = df['Order Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month
```

```python
df
```

```python
df['Category'].value_counts()
```

```python
df['Region'].value_counts()
```

```python
df['Segment'].value_counts()
```

```python
df['High Sale'] = df['Sales'] > df['Sales'].median()
```

```python
df.to_csv("F:\\DEPI\\Datasets\\Datasets\\cleaned_store_sales.csv", index=False)
```

# 1.1 Total Sales Overview 

```python
# Total Sales
total_sales = df['Sales'].sum()
print(f"Total Sales: ${total_sales:,.2f}")

# Basic sales statistics
print(df['Sales'].describe())
```

# 1.2 High-Sale Flag Distribution

```python
sales_by_region = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
sales_by_region.plot(kind='bar', title='Total Sales by Region')
plt.ylabel('Total Sales')
plt.show()
```

# 1.3 Average Sales by Delivery Time (Impact of Delivery Delays)

```python
# Group by Delivery Time
delivery_sales = df.groupby('Delivery Time (Days)')['Sales'].mean().reset_index()

# Plot
sns.lineplot(data=delivery_sales, x='Delivery Time (Days)', y='Sales', marker='o')
plt.title('Average Sales vs Delivery Time')
plt.xlabel('Delivery Time (Days)')
plt.ylabel('Average Sales ($)')
plt.grid(True)
plt.show()
```

# 2.1 Top-Performing Categories & Sub-Categories 

```python
# Total sales by category
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)

# Total sales by sub-category
subcategory_sales = df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False)

# Plotting
plt.figure(figsize=(10, 5))
category_sales.plot(kind='bar', color='skyblue')
plt.title('Total Sales by Category')
plt.ylabel('Sales ($)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

plt.figure(figsize=(12, 6))
subcategory_sales.plot(kind='bar', color='orange')
plt.title('Total Sales by Sub-Category')
plt.ylabel('Sales ($)')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()
```

# 2.2 Low-Performing Products

```python
# Products with lowest total sales
low_perf = df.groupby('Product Name')['Sales'].sum().sort_values().head(10)
print(low_perf)
```

# 2.3 Top-Spending Customers

```python
# Total sales per customer
top_customers = df.groupby(['Customer ID', 'Customer Name'])['Sales'].sum().sort_values(ascending=False).head(10)

# Display
print(top_customers)
```

# 2.4 Segment Performance

```python
# Total sales by segment
segment_sales = df.groupby('Segment')['Sales'].sum()

# Plot
segment_sales.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Sales by Customer Segment')
plt.ylabel('')
plt.show()
```

# 3.1 Monthly Sales Trend

```python
# Group by Year-Month
monthly_sales = df.groupby('Year-Month')['Sales'].sum()

# Plot
plt.figure(figsize=(14, 6))
monthly_sales.plot(kind='line', marker='o', color='teal')
plt.title('Monthly Sales Trend')
plt.xlabel('Year-Month')
plt.ylabel('Sales ($)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

# 3.2 Yearly Sales Growth

```python
# Group by Year
yearly_sales = df.groupby('Order Year')['Sales'].sum()

# Plot
yearly_sales.plot(kind='bar', color='coral')
plt.title('Total Sales by Year')
plt.xlabel('Year')
plt.ylabel('Sales ($)')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

# 3.3 Order Volume per Month

```python
monthly_orders = df.groupby('Year-Month')['Order ID'].nunique()

# Plot
monthly_orders.plot(kind='bar', figsize=(15, 6), color='mediumpurple')
plt.title('Number of Orders Per Month')
plt.xlabel('Year-Month')
plt.ylabel('Order Count')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
